```python
import math
import numpy as np
from Bio.PDB import PDBList, DSSP, HSExposureCB, PPBuilder
from Bio.PDB.PDBParser import PDBParser
import matplotlib.pyplot as plt
from matplotlib import patches
from Bio.SeqUtils import seq1
import pandas as pd
```

```python
rama_ss_ranges = [(-180, -180, 80, 60, 'E', 'blue'),
                  (-180, 50, 80, 130, 'E', 'blue'),
                  (-100, -180, 100, 60, 'P', 'green'),
                  (-100, 50, 100, 130, 'P', 'green'),
                  (-180, -120, 180, 170, 'H', 'red'),
                  (0, -180, 180, 360, 'L', 'yellow')]
```

```python
# Ramachandran regions
regions_matrix = []
with open(path + "ramachandran.dat") as f:
    for line in f:
        if line:
            regions_matrix.append([int(ele) for ele in line.strip().split()])

```

```python
#pdb_ids = [('2kkw', 'A')]
#for pdb_id, chain_id in pdb_ids:

pdb_id = '2kkw'
chain_id = 'A'

# Fetch a PDB file to the current dir
pdbl = PDBList()
pdbl.retrieve_pdb_file(pdb_id, pdir=path, file_format='pdb')  # Will save to pdbXXXX.ent

# Load the structure
structure = PDBParser(QUIET=True).get_structure(pdb_id, path + "pdb{}.ent".format(pdb_id))
```

```python
#Rename the file To .pdb
!mv "drive/MyDrive/SB_practical_notebooks/data/pdb2kkw.ent" "drive/MyDrive/SB_practical_notebooks/data/pdb2kkw.pdb"
```

```python
data = []
for model_index, model in enumerate(structure):
    print(model_index, model[chain_id])

    residues = [residue for residue in model[chain_id]]

    # WARNING DSSP calculates values only on the first model! 
    # You should save models into different files to get the differences
    dssp = DSSP(model, path + "pdb{}.pdb".format(pdb_id), dssp="./dssp-2.3.0/mkdssp")  # WARNING Check the path of mkdssp
    dssp = dict(dssp)

    hse = HSExposureCB(model[chain_id])
   
    ppb = PPBuilder()  # PolyPeptideBuilder
    rama_dict = {}
    for pp in ppb.build_peptides(model[chain_id]):
        phi_psi = pp.get_phi_psi_list()  # [(phi_residue_1, psi_residue_1), ...]
        for i, residue in enumerate(pp):
            phi, psi = phi_psi[i]
            ss_class = None
            if phi is not None and psi is not None:
                for x, y, width, height, ss_c, color in rama_ss_ranges:
                    if x <= phi < x + width and y <= psi < y + height:
                        ss_class = ss_c
                        break
            rama_dict[(chain_id, residue.id)] = [phi, psi, ss_class]

    for residue in residues:
        data.append((model_index, chain_id, "{}{}{}".format(*residue.id), dssp.get((chain_id, residue.id))[2],
                          dssp.get((chain_id, residue.id))[3],
                            *hse[(chain_id, residue.id)],
                          *rama_dict.get((chain_id, residue.id))))

data
```

```python
df = pd.DataFrame(data, columns=['model', 'chain', 'residue', 'dssp', 'rsa', 'hse_up', 'hse_down', 'hse_angle', 'phi', 'psi', 'ss_three'])
df
```

```python
df_agg = df.groupby('residue', as_index=False, sort=False).agg({'rsa': ['mean', 'std'],  'hse_up': ['mean', 'std'],  'hse_down': ['mean', 'std'], 'phi': ['mean', 'std'], 'psi': ['mean', 'std']})
df_agg
```

```python
fig, axes = plt.subplots(5, 1, figsize=(16, 24))

c = 0
for feature, new_df in df_agg.groupby(level=0, axis=1):
  if feature != 'residue':
    print(feature)
    new_df = new_df.droplevel(0, axis=1)  
    axes[c].errorbar(new_df.index.values, new_df['mean'], yerr=new_df['std'], ecolor='r', capthick=0.5)
    axes[c].set_title(feature)
    c += 1

plt.tight_layout()  # Remove figure padding
# plt.savefig(path + '{}_features.png'.format(pdb_id), bbox_inches='tight')
```
